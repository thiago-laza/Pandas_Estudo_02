Pegando um arquivo CSV e gerando um banco de dados SQL usando o SQLALCHEMY

```python
import sqlalchemy
```

```python
from sqlalchemy import create_engine, MetaData, Table, inspect, text
```

```python
engine = create_engine('sqlite:///estudantes.db')
```

```python
dados = '/home/laza/Dados/Alura/Pandas/c2_trabalhando_com_diferentes_formatos_arquivos/estudo/estudantes_csv_100.csv'
```

```python
df = pd.read_csv(dados)
```

```python
df.to_sql('estudantes', con=engine, index=False, if_exists='replace')
```

```python
df
```

```python
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```
