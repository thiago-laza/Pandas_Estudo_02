


Este DataFrame possui 8 colunas: id, name (nome), username (nome de usurio), email , address (endereo), phone(telefone), website e company (empresa). Note que as colunas address e company contm informaes aninhadas, ou seja, os dados dessas colunas esto organizados em subcampos. Por exemplo, na coluna address, voc encontrar subcampos como street (rua), suite (complemento), city (cidade) e zipcode (CEP). J na coluna company, os subcampos incluem name (nome da empresa), catchPhrase (slogan) e bs (rea de atuao). Isso significa que, para acessar essas informaes,  necessrio realizar o processo de normalizao.

O desafio agora  normalizar esse DataFrame, expandindo as colunas address e company em suas respectivas subcolunas para facilitar a visualizao e a anlise dos dados.

```python
import pandas as pd
import requests
```

```python
url = 'https://jsonplaceholder.typicode.com/users'
```

```python
response = requests.get(url)
```

```python
data = response.json()
```

```python
data
```

```python
df = pd.json_normalize(data)
```

```python
df
```

```python
df.to_json('dados_normalizados.json')
```
